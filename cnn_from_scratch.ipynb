- **transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1)**: I used this function to add random brightness, contrast, saturation and hue transformations to an image which helps the model generalize better to different lighting conditions, color distributions, and variations of the image appearance.



- **transforms.ToTensor()**: Turns a PIL Image or numpy array into a PyTorch tensor. PyTorch tensors are used in neural network training as data formats.



- **transforms.Normalize(mean, std)**: Normalizes tensor image with mean and standard deviation provided. When we normalize our data it helps us stabilize training and improve convergence speed. Where there is a function to claculate the mean and std of our images

<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__
**In Layer 1**

- I have used sixty-four filtering layers to pull out the basics of the RGB image.

- I the applied normalization on the sixty four feature maps to stabilize the network during training.

- The inclusion of non-linearity is needed at this stage for it to incorporate highly complex patterns so i used ReLU as my activation function.

- I then did a max pooling in order to decrease spatial dimensions by half, and concentrate on more important details.

- Lastly I added a dropout where some units are deactivated at random times so as to prevent overfitting issues from occurring.

- After 1st Conv + MaxPool: 112 x 112 with 64 channels as our input image was 224 X 224


**In Layer 2**

- I used 128 filters to enable capturing of fine-grained features that were not retained in layer 1.

- I again Normalized the feature maps for stable and faster training purposes.

- I went again with my activation function which converts input images into a mean value of zero and standard deviation with values less than one. 

- I applied Max Pooling again to further downsample the feature maps. This step reduced the spatial dimensions once more while preserving the essential features identified during the previous downsampling in Layer 1

- Dropout regularization is also applied again.

- After 2nd Conv + MaxPool: 56 X 56 with 128 channels



**In Layer 3**

- I used 256 filters to enable capturing of fine-grained features that were not retained in layer 2.

- I again Normalized the feature maps for stable and faster training purposes.

- I went again with my activation function which converts input images into a mean value of zero and standard deviation with values less than one. 

- I applied Max Pooling again to further downsample the feature maps. This step reduced the spatial dimensions once more while preserving the essential features identified during the previous downsampling in Layer 2

- Dropout regularization is also applied again which prevents overfitting.

- After 3rd Conv + MaxPool: 28 X 28 with 256 channels


**In Layer 4**

- I used 512 filters to enable capturing of fine-grained features that were not retained in layer 3.

- I again Normalized the feature maps for stable and faster training purposes.

- I went again with my activation function which converts input images into a mean value of zero and standard deviation with values less than one. 

- I applied Max Pooling again to further downsample the feature maps. This step reduced the spatial dimensions once more while preserving the essential features identified during the previous downsampling in Layer 3

- At this point in time, Dropout regularization is also applied again.

- After 4th Conv + MaxPool: 14 X 14 with 512 channels

**Fully Connected Layer**
- Flatten the output from the CNN layers

- I created a fully connected layer with an input of 512 * 14 * 14 just as the output from our 4th layer

- Batch Normalization is also useful for fully connected layers. BatchNorm1d is appropriate here because it normalizes across the features of the fully connected layer.

- Used The ReLU activation function which converts input images into a mean value of zero and standard deviation with values less than one again.

- Applied Dropout Regularization

- Then mapped the 4096 units to num_classes output units (one for each class).

- Then lastly I passed the tensor through the model created.

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

```python
batch_size = 64        # increased size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2       # fraction of the training data to reserve for validation
num_epochs = 75        # increased number of epochs for training
num_classes = 50       # number of classes. Do not change this
dropout = 0.3          # reduced dropout for our model
learning_rate = 0.0005  # Learning rate for SGD (or Adam)
opt = 'adam'            # optimizer. 'sgd' or 'adam' changed from sgd to adam
weight_decay = 0.05     # increased regularization. Increase this to combat overfitting   
```

```python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel

# get the data loaders using batch_size and valid_size defined in the previous
# cell
# HINT: do NOT copy/paste the values. Use the variables instead
data_loaders = get_data_loaders(batch_size, valid_size)

# instance model MyModel with num_classes and drouput defined in the previous
# cell
model = MyModel(num_classes, dropout)

# Get the optimizer using get_optimizer and the model you just created, the learning rate,
# the optimizer and the weight decay specified in the previous cell
optimizer = get_optimizer(
    model,
    opt,
    learning_rate,
    weight_decay,
)

# Get the loss using get_loss
loss = get_loss()


optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_lossa.pt",
    interactive_tracking=True
)
```
